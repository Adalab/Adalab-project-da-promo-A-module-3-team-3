```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import numpy as np
import pandas as pd


# Gráficos
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns


#  Gestión de warnings
# ------------------------------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")


# Codificación variables categóricas
# ------------------------------------------------------------------------------
from sklearn.preprocessing import OrdinalEncoder


# Modelado y evaluación
# ------------------------------------------------------------------------------
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import tree
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import GridSearchCV

pd.options.display.max_columns = None
```

```python
df = pd.read_pickle('../datos/bikes_casual.pkl')
```

```python
def ordinal_encoder1(dataframe, columna):
    
    ordinal = OrdinalEncoder(dtype = int)

    transformados_oe = ordinal.fit_transform(dataframe[[columna]])

    oe_df = pd.DataFrame(transformados_oe)

    oe_df.columns = ordinal.feature_names_in_

    columna += "_oe"
    
    dataframe[columna] = oe_df
     
    return dataframe
```

```python
categoricas = df.select_dtypes(include = "category")

for col in categoricas:

    df = ordinal_encoder1(df, col)
```
